```python
# analysis of the equivalent PUA terms 
# Read the text from a given file (csv or txt format)
with open('alternative_terms_of_PUA_en.csv', 'r', encoding='utf-8') as file:
    text = file.read()
 
# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords="PUA", colormap = "tab20").generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
```

```python
def join_row(row):
    # Join columns and replace spaces with a non-separator character (non-breaking space)
    return ' '.join(map(str, row)).replace(' ', '\u00A0')

# Load your CSV file
df = pd.read_csv('alternative_terms_of_PUA_en.csv')

# Process each row
df['united'] = df.apply(join_row, axis=1)

# Create one large string from all "united" words
united_text = ' '.join(df['united'])

# Generate and display the word cloud
wordcloud = WordCloud(width=800, height=400).generate(united_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
